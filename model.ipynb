```python
#getting the dataset

points=pd.read_csv('./data/data (1).csv')


x_col=points["34.84234527"]
y_col=points["34.13700585"]

n=len(x_col)





```

```python
x_array=np.array(x_col)
y_array=np.array(y_col)
##function to calcualte the error given by the actual minus expected 
##we take the derivative of this function in the gradient descent
def error_func():
    eror_sum=0
    m=0
    b=0
    


    for index in range(n):
        eror_sum=eror_sum+(y_array[index]-(m*x_array[index]+b))
    
    error=1/n*eror_sum

    return error

```

```python
def gradient_descent(m_now, b_now, points, L):
    m_gradient = 0
    b_gradient = 0
    n = float(len(points))
    for i in range(len(points)):
        x = points.iloc[i, 0]
        y = points.iloc[i, 1]
        m_gradient += -(2/n) * x * (y - (m_now * x + b_now))
        b_gradient += -(2/n) * (y - (m_now * x + b_now))
    m = m_now - L * m_gradient
    b = b_now - L * b_gradient
    return [m, b]
    

    



##run for loop calling this function multiple times to update m
```

```python
##number of epochs to train the gradeint descent function on

m = 0
b = 0
L = 0.0001
epochs = 1000

for i in range(epochs):
    m, b = gradient_descent(m, b,points, L)

print(m, b)




```
